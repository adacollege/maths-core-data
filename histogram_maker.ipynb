```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
```

```python
def make_hist(dataset, ax):
    hist_data = []
    edges = dataset["edges"]
    freqs = dataset["freqs"]
    for (l,u,f) in zip(edges[:-1],edges[1:],freqs):
        hist_data += [(l+u)/2 for _ in range(f)]
    sns.histplot(ax=ax, data = hist_data, bins=edges, stat='frequency')
    ax.set(ylabel='Frequency Density', xlabel=data_name)
    ax.set_title(dataset["series_name"])
    #ax.set_xticks(range(0,100,10))
    #ax.tick_params(left = True, bottom=True, which = "both")
    #ax.grid(which='minor')
    ax.set_xbound(0)
```

```python
data =[{"series_name":"Class A", "edges":[0, 20, 40, 50, 60, 65, 70, 85, 100], "freqs":[2, 3, 4, 6, 4, 6, 3, 1], "row":0, "col":0},
      {"series_name":"Class B", "edges":[0, 20, 40, 50, 55, 60, 70, 80, 90], "freqs":[1, 3, 6, 8, 7, 2, 2, 1], "row":0, "col":1}]
plt_name = 'student_scores'
data_name = 'Test Score'
```

```python
sns.set_style("darkgrid")
ncols = max([series["col"] for series in data])+1
nrows = max([series["row"] for series in data])+1
fig, axs = plt.subplots(nrows, ncols, figsize=(16,8), squeeze = False, sharex=True, sharey=True)
for ds in data:
    make_hist(ds, axs[ds["row"]][ds["col"]])
```

```python
fig.savefig('{}.png'.format(plt_name))
```
